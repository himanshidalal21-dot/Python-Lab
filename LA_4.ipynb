```python
import numpy as np

# Generate two 3x3 matrices with random integers (1 to 9)
A = np.random.randint(1, 10, (3, 3))
B = np.random.randint(1, 10, (3, 3))

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

# Addition
add = A + B
print("\nAddition (A + B):")
print(add)

# Multiplication
mul = np.dot(A, B)
print("\nMultiplication (A × B):")
print(mul)
```

```python
import numpy as np

print("Enter elements for 5x3 matrix:")

A = []
for i in range(5):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    A.append(row)

A = np.array(A)

print("\nEnter elements for 3x2 matrix:")

B = []
for i in range(3):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    B.append(row)

B = np.array(B)

# Multiply matrices
result = np.dot(A, B)

print("\nProduct Matrix (5x2):")
print(result)
```
