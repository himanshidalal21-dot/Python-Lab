```python
# Program to copy python file without comments

# Take file names from user
source = input("Enter source file name: ")
destination = input("Enter destination file name: ")

# Open source file
with open(source, "r") as file1:
    lines = file1.readlines()

# Remove comments and store clean code
clean_lines = []
for line in lines:
    line = line.strip()

    # Skip full line comments
    if line.startswith("#"):
        continue

    # Remove inline comments
    if "#" in line:
        line = line.split("#")[0].rstrip()

    # Add non-empty lines
    if line != "":
        clean_lines.append(line)

# Write cleaned content to destination file
with open(destination, "w") as file2:
    file2.write("\n".join(clean_lines))

# Print both files
print("\n--- Source File Content ---")
with open(source, "r") as file1:
    print(file1.read())

print("\n--- Destination File Content (No Comments) ---")
with open(destination, "r") as file2:
    print(file2.read())
```

```python
# Program to convert file content to uppercase

# Open source file in read mode
with open("demo.txt", "r") as file1:
    content = file1.read()

# Convert content to uppercase
upper_content = content.upper()

# Open new file in write mode
with open("output.txt", "w") as file2:
    file2.write(upper_content)

print("File copied successfully in uppercase.")
```
